```python
#2. Write a function to extract first word of s string (with many words separated by space).
def firstword(string):
  return string.split(' ',1)[0]

b = "Hello Everyone this is a string"
firstword(b)
```

```python
#3. Extract the first word from every string from a list of strings by using map function.
splitting = lambda x:x.split(' ',1)[0]
alist = ["Welcome to Python Programming",
         "This is a List of Strings",
         "Taking first word",
         "From Each String"]
result = map(splitting,alist)
for i in list(result):
  print(i)
```

```python
#4. Write a function to return a list of prime factors of a given number.
import math
def primefactors(n):
  while (n%2==0):
    print(2)
    n = n/2
  
  for i in range(3,int(math.sqrt(n)+1),2):
    while(n%i ==0):
      print(i)
      n = n/i
  if n>2:
    print(n)
primefactors(1050)
```

```python
#5. Write a function that finds 2nd largest among 4 numbers (Repetitions are allowed,without sorting).
def secondlarge(alist):
  alist.remove(max(alist))
  return max(alist)

print(secondlarge([1,2,3,4,5]))
print(secondlarge([100,134,412,541,599]))
```

```python
#6. Take an input from the user as string and find out all the non duplicate words from the input
#import re
#m = re.split(',| ',input())
m = input().replace(', ',' ').lower().split()
temp = []
for i in m:
  if i not in temp:
    temp.append(i)
print(' '.join(temp))
```

```python
#7. Take a input from the user and verify the data is a palindrome or not
m = input().upper()
if m == m[::-1]:
  print("Given String is a Palindrome...")
else:
  print("Not a Palindrome...")
```

```python
#8. Use exception handling to find out the LCM & HCF of two numbers
def HCF(num1,num2):
  return int((num1*num2)/LCM(num1,num2))
def LCM(num1,num2):
  if num1>num2:
    lcm = num1
  else:
    lcm = num2
  while True:
    if lcm%num1==0 and lcm%num2==0:
      break
    else:
      lcm += 1
  return lcm

a = int(input("Enter First Number : "))
b = int(input("Enter Second Number : "))
try:
  lcm = LCM(a,b)
  hcm = HCF(a,b)
  print(f"Least Common Multiple of {a,b} is : {lcm}")
  print(f"Highest Common Factor of {a,b} is : {hcm}")
except ZeroDivisionError:
  print(f"Numbers Cannot be Zero, {ZeroDivisionError} Recieved...!!!" )
```

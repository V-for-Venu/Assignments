```python
#6
from math import factorial as fact
n = 7
for i in range(n):
	for j in range(i+1):
		print(fact(i)//(fact(j)*fact(i-j)), end=" ")
	print()
```

```python
#13
m = 4
for i in range(1,6):
  for x in range(m):
    print(end='  ')
  m-=1
  print(*['*' for _ in range(0,i)])
m = 1
for i in range(4,0,-1):
  for _ in range(m):
    print(end='  ')
  m+=1
  print(*['*' for _ in range(i)])
```

```python
#15. 
def uscore(n):
  print(*['_' for _ in range(n*2)],end='',sep='')

def astr(n):
  print(*['*' for _ in range(n)],end='',sep='')

x = 8
for i in range(8):
  astr(x-i)
  uscore(i)
  astr(x-i)
  print()
```
